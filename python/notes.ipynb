**Importing all kinds of stuffs**

```python
#various imports etc.

import numpy as np
import datetime as dt
from datetime import date
from datetime import time
```

**Super basic math stuff in case my brain decides to go on a vacation**

```python
#math operations

x = np.random.randint(1,10)
y = np.random.randint(1,10)

print(x)
print(y)
print(f"{x}+{y}={x+y}")
print(f"{x}-{y}={x-y}")
print(f"{x}*{y}={x*y}")
print(f"{x}/{y}={x/y}")
print(f"{x}**{y}={x**y}")
print(f"{x}//{y}={x//y}")
print(f"{x}%{y}={x%y}")
```

**Lists**

```python
#filling a list with placeholder junk

myList = ['a','b','c','d',5]
x=0
while (x<5):
    myList.append(np.random.randint(1,10))
    x+=1
myList

myList[2:5] #lists are indexed from 0
myList[0:4:3] #third parameter is the step for the selected range. 3 selects every other element.
myList[::-1] #this returns the list in reverse order
myList.remove(5) #removal of specified element
myList.pop(1) #removal of n amount of elements from the end of the list
myList[-1] #last element of the list
myList[-2] #second last element of the list
```

**Tuples**

```python
myTuple = (1,2,3,4,5,3) #defining a tuple
myTuple.count(3) #counting how many times does 3 appear in the tuple
myTuple.index(3) #lowest index at which 3 occurs
myTuple[1:3] #selecting elements 1 to 3 from the tuple
myTuple[:4] #selecting first four elements from the tuple
```

**Dictionaries**

```python
#defining a dictionary

dictionary = {'one':1,
              'two':2,
              'three':3}

dictionary['three'] #calling an item from a dictionary
dictionary['three'] = 4 #reassigning different value to a key
dictionary['four'] = dictionary.pop('three') #replacing a key called 'three' with a key called 'four'
dictionary.items() #listing all the items of a dictionary
print(dictionary.keys()) #displaying all keys of a dictionary
print(dictionary.values()) #displaying all values of a dictionary
```

**Sets**

```python
#defining two sets

setA = {1,2,3,4}
setB = {5,6,7,8}

setA.union(setB) #adding two sets
setA.intersection(setB) #common part of two sets
setA.difference(setB) #returns the elements that arent common between two sets
```

**Datetime**

```python
#obtaining current date and time using datetime package

myDate = dt.datetime.now()
print(myDate)

#obtaining current date using datetime package

print(dt.date.today())

#it's all pretty self-explainatory, really

today = date.today()
day = today.day
month = today.month
year = today.year
print(day,month,year)

myTime = time(10,32,40)
hour = myTime.hour
minute = myTime.minute
second = myTime.second
print(hour,minute,second)

#wibbly-wobbly timey-wimey stuff (aka how to use time from datetime package)

print(time())
print(time(10, 32, 40))
print(time(hour=10, minute=32, second=40))
```
